### üß© Paso 1: Preprocesar el texto y generar bigramas

```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import defaultdict
import random
import string

nltk.download('punkt')

# Texto base
text = '''
I love natural language processing and I love machine learning.
Machine learning is a subset of artificial intelligence, and it is fascinating.
Natural language processing helps computers understand human language.
'''

# Preprocesamiento
text = text.lower()
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenizaci√≥n y bigramas
tokens = word_tokenize(text)
bigrams = list(ngrams(tokens, 2))

# Mostrar algunos bigramas
print("Ejemplo de bigramas:", bigrams[:10])
```

### üß© Paso 2: Construir un modelo estad√≠stico simple con bigramas

```python
model = defaultdict(list)
for w1, w2 in bigrams:
    model[w1].append(w2)

# Mostrar un ejemplo
print("Palabras que pueden seguir a 'machine':", model['machine'])
```

### üß© Paso 3: Generar una frase aleatoria con el modelo de bigramas

```python
word = random.choice(tokens)
sentence = [word]
for _ in range(15):
    next_words = model.get(word)
    if not next_words:
        break
    word = random.choice(next_words)
    sentence.append(word)

print("Frase generada:", ' '.join(sentence))
```

# Soluci√≥n (Click para revelar)

```python
# Escribe tu propio texto base:
texto_usuario = input("Introduce tu texto base: ")

# Preprocesamiento
texto_usuario = texto_usuario.lower()
texto_usuario = texto_usuario.translate(str.maketrans('', '', string.punctuation))

# Tokenizaci√≥n y bigramas
tokens = word_tokenize(texto_usuario)
bigrams = list(ngrams(tokens, 2))

# Crear modelo
modelo_usuario = defaultdict(list)
for w1, w2 in bigrams:
    modelo_usuario[w1].append(w2)

# Generar frase
if tokens:
    palabra = random.choice(tokens)
    oracion = [palabra]
    for _ in range(15):
        siguientes = modelo_usuario.get(palabra)
        if not siguientes:
            break
        palabra = random.choice(siguientes)
        oracion.append(palabra)
    print("Frase generada con tu texto:")
    print(' '.join(oracion))
else:
    print("No se pudo generar modelo con el texto proporcionado.")
```
